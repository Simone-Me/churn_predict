# 01 - Analyse exploratoire du churn client

Objectif : comprendre le dataset avant la modelisation. On regarde la taille, la cible, les valeurs manquantes et les premiers signaux metier.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

BASE_DIR = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
DATA_PATH = BASE_DIR / 'data' / 'customer_churn.csv'
df = pd.read_csv(DATA_PATH)
df.head()
```

## Taille du dataset

Le dataset contient les informations clients : profil, usage, support, paiement, satisfaction et la variable cible `churn`.

```python
print('Nombre de lignes :', df.shape[0])
print('Nombre de colonnes :', df.shape[1])
df.dtypes
```

## Analyse de la cible

Le churn est desequilibre : il y a beaucoup plus de clients non churners que de churners. C est pour cela que le recall est prioritaire.

```python
target_counts = df['churn'].value_counts().sort_index()
print(target_counts)
print((df['churn'].value_counts(normalize=True).sort_index() * 100).round(2))
target_counts.plot(kind='bar', title='Repartition de la cible churn')
plt.xlabel('churn')
plt.ylabel('nombre de clients')
plt.show()
```

## Valeurs manquantes

On verifie les colonnes incompletes. Ici, `complaint_type` peut etre vide : on l interprete comme une absence de plainte.

```python
missing = df.isna().sum().sort_values(ascending=False)
missing[missing > 0]
```

## Quelques comparaisons simples

On compare certains indicateurs selon la cible churn pour construire une intuition metier.

```python
cols = ['monthly_logins', 'last_login_days_ago', 'payment_failures', 'support_tickets', 'csat_score', 'nps_score', 'monthly_fee']
df.groupby('churn')[cols].mean().round(2)
```
